# Amylice: Reproducing Article Figures

This notebook reproduces result figures from the Amylice manuscript using local analysis outputs and figure-generation scripts.

Figures 1 and 2 are skipped because they are conceptual/workflow diagrams. The notebook starts from Figure 3 and uses a repeated structure: title, figure text, code, and the visible reproduced PNG figure.

```python
from pathlib import Path
import os
import subprocess
import sys

try:
    from IPython.display import Image, display
except Exception:
    Image = display = None

ROOT = Path.cwd()
BASE = ROOT / 'analysis_outputs' / 'amyloid_bc_control_vs_chili'

# Set to True to rerun figure-generation scripts before displaying figures.
# Figure 3 is regenerated from a processed technical-validation source table.
RUN_RECOMPUTE = False

os.environ.setdefault('MPLCONFIGDIR', '/private/tmp/matplotlib_codex_cache')
os.environ.setdefault('MPLBACKEND', 'Agg')


def reproduce_figure(image_path, script_name=None, width=1100):
    image = ROOT / image_path
    if script_name:
        script = ROOT / script_name
        print('Script:', script)
        if RUN_RECOMPUTE:
            subprocess.run([sys.executable, str(script)], cwd=ROOT, check=True)
    else:
        print('Script: generated figure or embedded local output')
    print('Figure:', image)
    if not image.exists():
        raise FileNotFoundError(image)
    if display is not None:
        display(Image(filename=str(image), width=width))
    else:
        print('Open image:', image)
```
